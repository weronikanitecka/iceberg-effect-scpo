```python
#Upload files
from google.colab import files
uploaded = files.upload()

# Build arrests_core.csv (individual-level, selected columns)
import pandas as pd

CORE_COLS = [
    'apprehension_date', 'apprehension_type', 'apprehension_state',
    'apprehension_aor', 'final_program', 'arresting_agency',
    'apprehension_method', 'apprehension_criminality', 'case_status',
    'case_category', 'departure_country', 'final_order_yes_no',
    'birth_year', 'citizenship_country', 'gender', 'departed_date',
    'final_order_date', 'apprehension_site_landmark', 'operation',
    'toa_current_duty_site', 'case_criminality', 'case_threat_level',
    'unique_identifier', 'apprehension_date_time',
]

print("Loading arrests (large file, ~5 min)...")
arrests = pd.read_excel('arrests-latest.xlsx', engine='openpyxl',
                         usecols=lambda c: c in CORE_COLS)
print(f"Loaded {len(arrests):,} rows")
arrests.to_csv('arrests_core.csv', index=False)
print("Saved arrests_core.csv")

# State-aggregated arrests
arrests_state = (
    arrests.groupby('apprehension_state')
    .size()
    .reset_index(name='ice_arrests')
)
arrests_state.to_csv('ice_arrests.csv', index=False)
print(f"State arrests: {len(arrests_state)} states")
print(arrests_state.sort_values('ice_arrests', ascending=False).head())

# State-aggregated detainers
print("Loading detainers (~5 min)...")
detainers = pd.read_excel('detainers-latest.xlsx', engine='openpyxl')
print(f"Loaded {len(detainers):,} rows")
det_state = (
    detainers.groupby('facility_state')
    .size()
    .reset_index(name='ice_detainers')
)
det_state.to_csv('ice_detainers.csv', index=False)
print(f"State detainers: {len(det_state)} states")

# Download all CSVs
files.download('arrests_core.csv')
files.download('ice_arrests.csv')
files.download('ice_detainers.csv')
print("Done. Place files in data/raw/")
```
